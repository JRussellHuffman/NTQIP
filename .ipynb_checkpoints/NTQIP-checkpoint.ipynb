```python
%matplotlib inline

# Importing standard Qiskit libraries and configuring account
from qiskit import QuantumCircuit, execute, Aer, IBMQ

IBMQ.load_account()
provider = IBMQ.get_provider('ibm-q-internal', 'deployed', 'default')


from math import pi
import qiskit.quantum_info as qi
```

```python
from PIL import Image

im = Image.open('data/bee-source.png')
pix = im.load()
print (im.size)
print (im.width * im.height)

imageList = []
maxListLength = 127 #this is also the number of qubits
```

```python
def fillPixelList(colorChannel):
    imageList.append([])
    currentIndex = 0
    listIndex = len(imageList)-1
    for i in range(im.width):
        for j in range(im.height):
            if (currentIndex >= maxListLength):
                currentIndex = 1
                imageList.append([])
                listIndex += 1
            else:
                currentIndex += 1  

            pixelValue = pix[i,j]
            imageList[listIndex].append(pixelValue[colorChannel])
#     print(imageList)
```

```python
def runNTQIP(channelName, backendName, textName):
    if channelName == "red":
        fillPixelList(0) #red
    elif channelName == "green":
        fillPixelList(1) #green
    elif channelName == "blue":
        fillPixelList(2) #blue
    circuitList = []

    for i in range(len(imageList)):
        qc = QuantumCircuit(len(imageList[i]))

        for index, item in enumerate(imageList[i]):
#             qc.rx((1-(item/255))*pi,index) # black returns a dot
            qc.rx(((item/255))*pi,index) # white returns a dot
    
            #---- add quantum gates here ---- #
        
            qc.x(index) # example: the x gate inverts the image
        
            #-------------------------------- #

        qc.measure_all()
        circuitList.append(qc)
        
    if backendName == "sim":
        backend = Aer.get_backend('qasm_simulator')
    else :
        backend = provider.get_backend(backendName)

    shots = 100
    job = execute(circuitList, backend, shots=shots, memory=True)

    result = job.result()

    def storeOutputs(circuitMemory):
        # need an array to drop all the results into
        outputArray = []

        for i in range(0, shots):
            outputArray.append(memory[i])

        return outputArray

    finalOutput = []
    for i in range(len(circuitList)):
        memory = result.get_memory(i)
        storeOutputs(memory)
        finalOutput.append(storeOutputs(memory))

    finalfinal = sum(finalOutput, [])

    #somehow print this value to a text file.
    textfile = open("data/"+ textName + "-" + backendName + "-" + str(maxListLength) + ".txt", "w")
    for element in finalfinal:
        textfile.write(element + ",")
    textfile.close()
    
    return finalfinal
```

```python
QuantumImageList = runNTQIP("blue", "sim", "bee-test")
```

```python
# Imports PIL module
import PIL
 
# creating a new image object with all white pixels
outputImage = PIL.Image.new(mode="RGB", size=(im.width, im.height), color=(255,255,255))



pixelMap = [0] * len(QuantumImageList[0])

print(len(pixelMap))

 
for i in range(len(QuantumImageList)):
    for j in range(len(QuantumImageList[i])):
        pixelMap[j] += int(QuantumImageList[i][j])

#correct pixel brightness and draw in the correct place
for i in range(len(pixelMap)):

    xCoordinate = int(i/im.width)
    yCoordinate = i - xCoordinate*im.width
    
    pixelMap[i] = int(pixelMap[i]/len(QuantumImageList) * 255)
    
    print(pixelMap[i])
    outputImage.putpixel((xCoordinate, yCoordinate), (pixelMap[i], pixelMap[i], pixelMap[i]))



finalImage = outputImage.rotate(180).show()
finalImage = outputImage.save("data/quantum-image-1-x.jpg")
```
